```python
df[(df['is_canceled'] == 0) & (df['arrival_date_year'] == 2017)]\
.groupby('arrival_date_month', as_index = False)\
.agg({'is_canceled' : 'count'}).sort_values('is_canceled', ascending = False)\
.head(1)
# Самый популярный месяц в 2017 году с успешными бронированиями - Май
# Следовательно, месяц поменялся
```

```python
df[(df['hotel'] == 'City Hotel') & (df['is_canceled'] == 1)]\
.groupby(['arrival_date_year', 'arrival_date_month'], )\
.agg({'is_canceled' : 'count'}).sort_values(['arrival_date_year','is_canceled'], ascending = False)
# Самыми часто отменяемыми месяцами в 2015, 2016 и 2017 годах стали - Сентябрь, Октябрь и Май, соответственно.
```

### Посмотрите на числовые характеристики трёх переменных: adults, children и babies. Какая из них имеет наибольшее среднее значение?

```python
adults = df.adults.mean()
children = df.children.mean()
babies = df.babies.mean()
dic = {adults : 'adults', children : 'children', babies : 'babies'}
print(f'Наибольшее среднее значение имеет переменная {dic[max(dic)]} ', round(max(dic), 3))
# При выводе, в принципе, все сказано
# (Да и лень смотреть на три строчки, искать наибольшее значение, лучше пусть выйдет одна и скажет что больше) 
```

### Создайте колонку total_kids, объединив children и babies. Отели какого типа в среднем пользуются большей популярностью у клиентов с детьми?

```python
# Не очень понял задание. Я считаю некорректной формулировку 'в среднем пользуются'. В данном задании, я считаю,
# стоит рассматривать только семьи с детьми и показать в процентах, куда какая доля семей с детьми уехала.

df['total_kids'] = df.children + df.babies
df_01 = df[df['total_kids'] > 0].groupby('hotel').agg({'total_kids' : 'count'})
df_01['total_kids'] = (df_01['total_kids'] / df_01['total_kids'].sum()).round(2)
df_01
# Получается, если у клиента есть дети, то он скорее выберет Городской отель, а не Курортный отель. 
```

### Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. Посчитайте отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах (churn rate). Укажите, среди какой группы показатель выше.

```python
df['has_kids'] = df['total_kids'].astype('bool')
```

```python
df['num_of_people'] = df.total_kids + df.adults
num_of_res_people = df.groupby('reservation_status').agg({'num_of_people' : 'sum'})
num_of_res_people['churn_rate'] = (num_of_res_people.num_of_people / df.num_of_people.sum()).round(2)
```
