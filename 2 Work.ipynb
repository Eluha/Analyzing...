```python
taxi.groupby('borough', as_index = False)\
.aggregate({'pickups' : 'sum'})\
.sort_values(by = 'pickups', ascending = False)
```

```python
group_taxi = taxi.groupby('borough', as_index = False)\
.aggregate({'pickups' : 'sum'})\
.sort_values(by = 'pickups', ascending = False)

result = group_taxi.pickups.sum()
```

10) Продолжим изучение данных и посмотрим на число поездок в выходные дни. Сгруппируйте данные по двум признакам: району города и является ли день выходным (колонки borough и hday). Сравните среднее число поездок, и назовите районы, из которых по праздникам в среднем поступает больше заказов, чем в обычные дни.

```python
# Этот кусок кода будет работать только в том случае, если каждому району будут сопоставлены данные о праздничном дне или нет. 
#Если, например, у Бронкса не было бы данных о праздничных днях, то кряк. Но данные выводятся норм. 
taxi_new_df, i = taxi.groupby(['borough', 'hday'], as_index = False)\
                     .aggregate({'pickups' : 'mean'}), 0

for i in range(0, len(taxi_new_df.pickups) - 1 , 2):
    if taxi_new_df.pickups[i] < taxi_new_df.pickups[i+1]:
        print(taxi_new_df.borough[i])
```
